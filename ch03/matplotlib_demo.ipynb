```python
import numpy as np
import matplotlib.pyplot as plt
def f(t):
    return np.exp(-t) * np.cos(2 * np.pi * t)
a = np.arange(0.0, 5.0, 0.02)
plt.subplot(211)
plt.plot(a, f(a))
plt.subplot(2, 1, 2)
plt.plot(a, np.cos(2 * np.pi * a), 'r--')
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
matplotlib.rcParams['font.family']='SimHei'
matplotlib.rcParams['axes.unicode_minus']=False
matplotlib.rcParams['font.size']=20
a = np.arange(0.0,5.0,0.02)
plt.title('')
plt.xlabel(' ')
plt.ylabel(' ')
plt.plot(a,np.cos(2*np.pi*a),'r--')
plt.show()
```

```python
import matplotlib.pyplot as plt
import random

plt.rcParams['font.sans-serif'] = ['KaiTi']  # 
plt.rcParams['axes.unicode_minus'] = False  # 
plt.figure()
month_days = 31
x = [i for i in range(1, month_days)]
y = [random.randrange(20, 41) for i in range(1, month_days)]

plt.bar(x, y, color="r", label="")
plt.title("")
plt.xlabel("()")
plt.ylabel("()")
plt.legend()
# plt.xticks(x)
# plt.yticks([i for i in range(0, 41)])
plt.show()
```

```python
import matplotlib.pyplot as plt
import random
plt.rcParams['font.sans-serif'] = ['KaiTi']  # 
plt.rcParams['axes.unicode_minus'] = False  # 
plt.figure()
month_days = 5
x = [i + 1 for i in range(1, month_days + 1)]
y = [random.randrange(20, 41) for i in range(1, month_days + 1)]
plt.bar(x, y, label="", color=["r", "y", "b", "g", "c"])
plt.xlabel("()")
plt.ylabel("()")
plt.legend()
plt.xticks(x, ["{0}".format(i + 1) for i in range(0, month_days + 1)])
plt.show()
```
